```python
#Plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df["generated_power_kw"], bins=30, kde=True, color='purple')
plt.title("Distribution of Generated Power (kW)")
plt.xlabel("Generated Power(kW)")
plt.ylabel("Frequency")
plt.show()
```

Bivariate Analysis with Scatter Plot

```python
# Scatter Plot with Target Feature
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column], df['generated_power_kw'], edgecolors="black")
    plt.title(f"{column} vs Generated Power(kW)")
    plt.xlabel(column)
    plt.ylabel('Generated Power(kW)')
plt.tight_layout()
plt.show()
```

```python
# Correlation
df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap="coolwarm", annot=True,fmt='.2f') # annot displays the values in the boxes
plt.show()
```

Managing Outliers

```python
# Outlier
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.tight_layout()
plt.show()
```
